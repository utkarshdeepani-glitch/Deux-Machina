```python
#################Parameters#######################
VY_THRESHOLD_HARD = -2
VY_THRESHOLD_SOFT = -0.7 # For close to ground check
ANGLE_THRESHOLD_SOFT = 0.05 # For close to ground check
ANGLE_THRESHOLD_HARD = 0.2
VX_DEADZONE = 0.8
X_DEADZONE = 0.3
##################################################

# Initialise the environment
env = gym.make("LunarLander-v3", render_mode="human")

# Reset the environment to generate the first observation
observation, info = env.reset(seed=42)
x, y, vx, vy, angle, ang_vel, leg1, leg2 = observation
run = True
total_reward = 0
while(run):
    # this is where you would insert your policy
    if vy < VY_THRESHOLD_HARD:
        action = 2
    elif abs(angle) > ANGLE_THRESHOLD_HARD:
        action = 1 if angle < 0 else 3
    elif abs(x) > X_DEADZONE:
        action = 1 if x > 0 else 3
    elif vy < VY_THRESHOLD_SOFT:
        action = 2
    elif abs(vx) > VX_DEADZONE:
        action = 1 if vx > 0 else 3
    elif abs(angle) > ANGLE_THRESHOLD_SOFT:
        action = 1 if angle < 0 else 3
    else:
        action = 0

    # step (transition) through the environment with the action
    # receiving the next observation, reward and if the episode has terminated or truncated
    observation, reward, terminated, truncated, info = env.step(action)
    x, y, vx, vy, angle, ang_vel, leg1, leg2 = observation
    total_reward += reward
    # If the episode has ended then we can reset to start a new episode
    if terminated or truncated:
        observation, info = env.reset()
        x, y, vx, vy, angle, ang_vel, leg1, leg2 = observation
        run = False

print("Total Reward:", total_reward)
env.close()
```
